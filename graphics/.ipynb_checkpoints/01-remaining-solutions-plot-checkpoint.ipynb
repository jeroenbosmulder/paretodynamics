```python
import numpy as np

def calculate_indicators(filename):
    sudoku = np.genfromtxt(filename, delimiter=',',dtype=int, missing_values='',filling_values=0)
    sudoku = (sudoku>0).astype(int)
    instance = []
    for i in range(81):
        x,y = divmod(i,9)
        entry = sudoku[3*x:(3*x+3),3*y:(3*y+3)].flatten()
        instance.append(entry/sum(entry))
    instance = np.asarray(instance)
    return [get_energy(instance)[0], get_entropy(instance)]
```

```python
t=np.append(float('inf'),np.diff(_energies)/np.diff(_entropies))
threshold = 0.606
idx = np.where(np.logical_and(t>threshold-0.01, t<=threshold))[0][0]
slope=t[idx]
point = (_entropies[idx+1],_energies[idx+1])
```

```python
text = "$S-E/T={0:.2f}$".format(point[0]-point[1]/slope)

fig, ax = plt.subplots(figsize=(19, 10))
plt.axvline(x=0, color="gray", linestyle="dotted")    
plt.axhline(y=0, color="gray", linestyle="dotted")    
plt.title('REMAINING SOLUTIONS PLOT')
plt.xlabel('logsize(set of remaining solutions)')
plt.ylabel('quality(set of remaining solutions)')

markersize = 200
plt.plot(_entropies,_energies, linewidth=4, color=orange, 
            label='Pareto front'.format(len(_entropies)),zorder=-1)
plt.scatter(_entropies[idx],_energies[idx], marker='D',s=markersize,color='dodgerblue')
plt.scatter(_entropies[idx],_energies[idx], marker='D',s=[100],color='dodgerblue',
            label='Paerto slope $T={0:.2f}$'.format(threshold))
plt.axline(xy1=point,slope=slope,lw=3,color='dodgerblue', 
            label='Tangent line: {}'.format(text))
plt.text(point[0]+2.5,point[1]-2.5,text,
            bbox=dict(facecolor='none', edgecolor='dodgerblue', pad=5))
plt.fill_between(np.append(_entropies,0),np.append(_energies,0), max(_energies), color=orange, alpha=0.1,
             label='Sets of Remaining Solutions')
handles, labels = plt.gca().get_legend_handles_labels()
order = [3, 0, 2, 1]

plt.legend([handles[idx] for idx in order], [labels[idx] for idx in order],loc='lower right',facecolor=legend_color)
plt.savefig(presentation_dir.format(file_name='01-remaining-solutions-plot-06'),transparent=True)
```

```python
text_halfplane = "$S-E/T<{0:.2f}$".format(point[0]-point[1]/slope)

fig, ax = plt.subplots(figsize=(19, 10))
plt.axvline(x=0, color="gray", linestyle="dotted")    
plt.axhline(y=0, color="gray", linestyle="dotted")    
plt.title('REMAINING SOLUTIONS PLOT')
plt.xlabel('logsize(set of remaining solutions)')
plt.ylabel('quality(set of remaining solutions)')

markersize = 200
plt.plot(_entropies,_energies, linewidth=4, color=orange, label='Pareto front'.format(len(_entropies)),zorder=-1)
plt.scatter(_entropies[idx],_energies[idx], marker='D',s=markersize,color='dodgerblue')
plt.axline(xy1=point,slope=slope,lw=3,color='dodgerblue',
            label='Tangent line: {}'.format(text))
plt.text(point[0]+2.5,point[1]-2.5,text,
            bbox=dict(facecolor='none', edgecolor='dodgerblue', pad=5))
plt.text(point[0]+2.5,point[1]+50,text_halfplane,
            bbox=dict(facecolor='none', edgecolor='dodgerblue', pad=5))

xlimits = ax.get_xlim()
ylimits = ax.get_ylim()
x = np.linspace(-50,150, 1000)
y_line = slope * (x - point[0]) + point[1]


plt.fill_between(x, y_line, max(_energies), color='lightblue', alpha=0.2, 
            label='Half-plane: {}'.format(text_halfplane))
plt.legend(loc='lower right',facecolor=legend_color)
ax.set_xlim(xlimits)
ax.set_ylim(ylimits)
plt.savefig(presentation_dir.format(file_name='01-remaining-solutions-plot-07'),transparent=True)
```
